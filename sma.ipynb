```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader as pdr
%matplotlib inline
```

```python
dji = pdr.get_data_yahoo('^DJI', '2019-06-10')
dji.head()
```

```python
dji_close = pd.DataFrame(dji.Close)
```

```python
dji_close['MA_7'] = dji_close.Close.rolling(7).mean()
dji_close['MA_30'] = dji_close.Close.rolling(30).mean()
```

```python
plt.figure(figsize=(15,8))
plt.grid(True)
plt.plot(dji_close['Close'],label='DJI')
plt.plot(dji_close['MA_7'],label='MA 7 days')
plt.plot(dji_close['MA_30'],label='MA 30 days')
plt.legend(loc = 2)
```
